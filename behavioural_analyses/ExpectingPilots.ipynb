```python
def duration(data):
    # duration in minutes 
    total_duration = (data.iloc[-1, :].time_second - data.iloc[0, :].time_first) / 60

    # how much of that was break time?
    # how long from the last trial in one block to the first trial in the next block?
    break_times = []
    for i in range(len(data)-1):
        if data.iloc[i, :].block != data.iloc[i+1, :].block:
            break_time = data.iloc[i+1, :].time_first - data.iloc[i, :].time_second
            break_times.append(break_time)


    break_time = sum(break_times) / 60  # in minutes
    active_time = total_duration - break_time

    print(f"Total duration: {total_duration:.2f} minutes")
    print(f"Break time: {break_time:.2f} minutes") # talked a lot with line hehe
    print(f"Active time: {active_time:.2f} minutes")
    print(f"Active time per block: {active_time / data['block'].nunique():.2f} minutes")


def duration_estimation(n_blocks, n_events_per_block, ISI, rng_interval, avg_response_time=0.5):
    # estimated duration in minutes
    avg_rng = sum(rng_interval) / 2
    estimated_duration = n_blocks * (n_events_per_block * (ISI + avg_rng + avg_response_time) ) / 60
    print(f"Estimated duration: {estimated_duration:.2f} minutes")
```

# Check that motor preparation is balanced across conditions

```python
conditions = pd.DataFrame()

for finger1 in ["index", "middle"]:
    for rep in ["repeated", "unrepeated"]:
        finger2 = finger1 if rep == "repeated" else ("middle" if finger1 == "index" else "index")
        for exp in ["expected", "unexpected"]:
            
            motor_prep = finger2 if exp == "expected" else ("middle" if finger2 == "index" else "index") 
            # do something

            tmp_data = {
                "finger1": finger1,
                "finger2": finger2,
                "rep": rep,
                "exp": exp,
                "motor_prep": motor_prep
            }
            conditions = pd.concat([conditions, pd.DataFrame([tmp_data])], ignore_index=True)
```

```python
conditions
```

if we want to compare expected vs unexpected trials do we have equal trials with motor preparation for index and middle finger? and if we want to compare rep vs non rep trials do we have equal trials with motor preparation for index and middle finger?

*Note:* one could argue that after the second stimulus there is a difference between expected and unexpected trials in terms of motor preparation, since in unexpected trials the participant needs to change their prepared response. We can potentially argue that the motor preparation and stimulus response should happen in the two different hemispheres (stimulus is presented to the left hand, response with the right hand), but still something to consider.

```python
print("\nMotor preparation counts for expected vs unexpected trials:")
print(conditions.groupby("exp")["motor_prep"].value_counts())

print("\nMotor preparation counts for repeated vs non-repeated trials:")
print(conditions.groupby("rep")["motor_prep"].value_counts())

print(conditions.groupby(["exp","rep"])["motor_prep"].value_counts())
```
